```python
import pandas as pd
import numpy as np
```

```python
data_filename = "Daily Update Data Agregat Covid-19 Jakarta.xlsx"
```

```python
dataset = []
```

```python
for i in range (0,10): 
    sheet = pd.read_excel(data_filename,sheet_name=i)
    dataset.append(sheet)
```

# Dataset variable information 
- **DJ** = Data Jakarta
- **DIJ** = Data Indonesia and Jakarta
- **Suspek** = Data Suspek
- **RDT ab** = Data  RDT ab
- **RDT ag** = Data RDT ag
- **HL** = Data Hasil Lab
- **V** = Data vaksinasi
- **P** = Data pemakaman
- **VW** = Data vaksinasi wilayah
- **DVWI** = Daily vaksinasi wilayah

```python
df_DJ = dataset[0]
df_DIJ = dataset[1]
df_Suspek = dataset[2]
df_RDTab = dataset[3]
df_RDTag = dataset[4]
df_HL = dataset[5]
df_V = dataset[6]
df_P = dataset[7]
df_VW = dataset[8]
df_DVW = dataset[9]
```

```python
idx = 0 
for i in dataset : 
    print(f"dataset ke dari excel {idx}: ")
    print(i.info())
    print()
    idx+=1
```
